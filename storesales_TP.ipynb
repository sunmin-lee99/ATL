```python
# import module
import numpy as np
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
# load data
train = TabularDataset("/content/drive/MyDrive/kaggle/store-sales-time-series-forecasting/train.csv")
test = TabularDataset("/content/drive/MyDrive/kaggle/store-sales-time-series-forecasting/test.csv")
```

```python
df = pd.concat([train,test],axis=0)

id, label = "id", "sales"
eval_metric = "accuracy"

num_columns = []
for col in train:
    if train[col].dtypes != "object" and col != label:
        num_columns.append(col)
```

```python
# training model
models = TabularPredictor(label=label, eval_metric=eval_metric).fit(train.drop(columns=[id]))
```

```python
# prediction
pred = models.predict(test.drop(columns=[id]))
```

```python
# submission
submission = pd.read_csv("/content/drive/MyDrive/kaggle/store-sales-time-series-forecasting/sample_submission.csv")

submission[label] = pred
submission.to_csv("submission2.csv", index=False)
```
